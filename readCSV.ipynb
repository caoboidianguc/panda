```python
import pandas as pd

file = pd.read_csv('Messages.csv')
file.head()

#in Messages.csv has a note at the end of its file which is not a right type of csv, be aware of that
```

```python
file.loc[:,['Direction']].head()
```

```python
suaIndex = file.copy()
suaIndex['NgayThang'] = suaIndex.index


suaIndex.head()
```

```python

suaIndex['Type'] = suaIndex.Direction
suaIndex['Direction'] = suaIndex.Number
suaIndex['Number'] = suaIndex.Destination
suaIndex['Destination'] = suaIndex.Time
suaIndex['Time'] = suaIndex.Date
#cause suaIndex had Date all wrong, so we get to delete it
suaIndex.reset_index(drop=True, inplace=True)
suaIndex['Type'].head()
```

```python
xoaDate = suaIndex.copy()
del xoaDate['Date']
xoaDate.columns = ['Time','Destination','Number','Direction','Type','NgayThang']
print(f'Tong cong co {len(xoaDate)} dong')
xoaDate.head()
```
