```python
data = {'State': ['Ohio','Ohio','Ohio','Nevada','Nevada','Nevada'],
'Year': [2000,2001,2002,2001,2002,2003],
'Population': [1.5,1.7,3.6,2.4,2.9,3.2]}

xuly = pd.DataFrame(data)
xuly.head()
```

```python
import numpy as np

xuly2 = xuly.copy()
xuly2['debt'] = np.arange(6.)
#reorder all columns
xuly2.columns = ['Debt','Year','Population','State']
xuly2.index = ['mot','hai','ba','bon','nam','sau']
xuly2
```

```python
gia = ['low','low','high','medium','high']
#if we don't specific indexing for each values in 'gia' for giaTri out put will be NaN
giaTri = pd.Series(gia, index=['mot','hai','ba','bon','sau'])

xuly2['Nhan Xet'] = giaTri
xuly2.loc[:,('Population','State')].head()
```

```python
xuly2.loc[:,['Debt','Year']]
```

```python
#delete in dataFrame
#by default .drop() have axis=0 which is row, we can change this by set axis=1 to indicating column
xoaYear = xuly2.drop('Year',axis=1)
print(xoaYear)
print('----------Year still in xuly2-------')
print(xuly2)
```

```python
xuly2['TBinh2'] = xuly2.Population < 2.5
xuly2
```

```python
#the other parameter is "inplace". if this set to True, DataFrame will update insteal of return a  copy
xoaTT = xuly2.copy()
print(xoaTT)
#set axis=1 to a column
xoaTT.drop('Year', inplace=True, axis=1)
print(xoaTT)
```

```python
#another way to delete a column
del xoaTT['Nhan Xet']
xoaTT
```
